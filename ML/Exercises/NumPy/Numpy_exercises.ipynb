```python
class M:
    def __init__(self, mat):
        self.mat = mat

    def __matmul__(self, other):
        return M(self.mat @ other.mat)
```

```python
m1 = M(np.random.randint(1, 10, (3, 3)))
m2 = M(np.random.randint(1, 10, (3, 3)))
```

```python
m1.mat
```

```python
m2.mat
```

```python
m3 = m1 @ m2
```
