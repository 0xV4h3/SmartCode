```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification, make_blobs, make_moons

from sklearn.mixture import GaussianMixture
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import SpectralClustering
```

# Dataset 1 

```python
X1, y1 = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=10)

plt.scatter(X1[:, 0], X1[:, 1], s=10, c=y1);
```

```python
X1.shape, y1.shape
```

```python
kmeans1 = KMeans(n_clusters=2, init='random', n_init=10)
p_kmeans_1 = kmeans1.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], c=p_kmeans_1, s=50, cmap='viridis')
```

```python
agglomerative1 = AgglomerativeClustering(n_clusters=2, linkage='ward', distance_threshold=None)
p_agglomerative_1 = agglomerative1.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], c=p_agglomerative_1, s=50, cmap='viridis')
```

```python
gmm1 = GaussianMixture(n_components=2)
p_gmm_1 = gmm1.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], c=p_gmm_1, s=50, cmap='viridis')
```

```python
spectral1 = SpectralClustering(n_clusters=2)
p_spectral_1 = spectral1.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], c=p_spectral_1, s=50, cmap='viridis')
```

```python
db1 = DBSCAN(eps=0.3, min_samples=30)
p_db_1 = db1.fit_predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], c=p_db_1, s=50, cmap='viridis')
```
