```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_moons, load_iris

from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN
```

# Aglomerative Clustering

```python
X, y_true = make_blobs(n_samples=200, centers=4, cluster_std=0.7, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
model = AgglomerativeClustering(n_clusters=None, linkage='ward', distance_threshold=7)
p = model.fit_predict(X)
# p
plt.scatter(X[:, 0], X[:, 1], c=p, s=50, cmap='viridis');
```

```python
model = AgglomerativeClustering(n_clusters=4, linkage='average', distance_threshold=None)
p1 = model.fit_predict(X)
# p1
plt.scatter(X[:, 0], X[:, 1], c=p1, s=50, cmap='viridis');
```

# GMM

```python
X, y_true = make_blobs(n_samples=200, centers=4, cluster_std=0.7, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
gmm_sk = GaussianMixture(n_components=4)
gmm_sk.fit(X)
preds_sk = gmm_sk.predict(X)
# sk_scores = gmm_sk.score_samples(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=preds_sk, s=50, cmap='viridis');
```

### Iris data

```python
iris = load_iris()
X_iris = iris.data
X_iris.shape
```

```python
y = iris.target
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=y, s=50, cmap='viridis');
```

```python
gmm_sk_iris = GaussianMixture(n_components=3)
gmm_sk_iris.fit(X_iris)
preds_sk_iris = gmm_sk_iris.predict(X_iris)
```

```python
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=preds_sk_iris, s=50, cmap='viridis');
```

# DBSCAN

### Blobs

```python
X, y_true = make_blobs(n_samples=200, centers=4, cluster_std=0.7, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
model_db = DBSCAN(eps=0.73, min_samples=3)
y_pred_db = model_db.fit_predict(X)

scatter = plt.scatter(X[:, 0], X[:, 1], c=y_pred_db, s=50, cmap="viridis")
plt.legend(*scatter.legend_elements());
```

### Moons

```python
X_m, y_true = make_moons(n_samples=300, noise=0.1, shuffle=True, random_state=0)
plt.scatter(X_m[:, 0], X_m[:, 1], s=50);
```

```python
model_dm = DBSCAN(eps=0.17, min_samples=5)
y_pred_dm = model_dm.fit_predict(X_m)

scatter = plt.scatter(X_m[:, 0], X_m[:, 1], c=y_pred_dm, s=50, cmap="viridis")
plt.legend(*scatter.legend_elements());
```
