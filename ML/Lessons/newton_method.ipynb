```python
import numpy as np
```

```python
def newton(func, dfunc, x0=1, it=100, th=1e-15):
    # while abs(func(x0)) > th:
    for i in range(it):
        df = dfunc(x0)
        if df == 0:
            break
        x0 = x0 - func(x0) / df
        # print(x0)
    else:
        return x0
```

```python
def f1(x):
    return x ** 2 - 7 * x + 10

def df1(x):
    return 2 * x - 7
```

```python
# newton(f1, df1, x0=3)
```

```python
def all_solutions(func, dfunc, th=1e-15, it=50, x_min=-10, x_max=10, step=0.1, round_=14):
    sol = set()
    for x0 in np.arange(x_min, x_max, step):
        x = newton(func, dfunc, x0=x0, th=th, it=it)
        if x is not None:
            sol.add(float(np.round(x, round_)))
    return list(sol)
```

```python
sols = all_solutions(f1, df1)
sols
```

```python
def f2(x):
    return x ** 3 - 4 * x ** 2 - 4 * x + 16

def df2(x):
    return 3 * x ** 2 - 8 * x - 4
```

```python
all_solutions(f2, df2)
```

```python
def f3(x):
    return x ** 5 - 4 * x ** 3

def df3(x):
    return 5 * x ** 4 - 12 * x ** 2
```

```python
sols = all_solutions(f3, df3, it=100)
sols
```
