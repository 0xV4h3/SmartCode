```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = np.array([3, 5, 4, 6])
weigth = np.array([8, 16, 15, 20])
```

```python
age_reshaped = age.reshape(-1, 1)
age_reshaped
```

```python
age_ones = np.ones(len(age)).reshape(-1, 1)
age_ones
```

```python
x = np.concatenate((age_ones , age_reshaped), axis = 1)
```

```python
x
```

```python
beta = np.linalg.inv((x.T @ x)) @ x.T @ weigth
beta
```

```python
b, k = beta
```

```python
def f(x):
    return k * x + b
```

```python
f(22)
```

```python
plt.scatter(age, weigth, color="r")
plt.plot(age, f(age), marker=".", markersize=15)
plt.grid()
```
