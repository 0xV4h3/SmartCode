```python
def mean(x):
    return x.sum() / len(x)
```

```python
def var(x):
    # return ((x - mean(x)) ** 2).sum() / len(x)
    return mean(x ** 2) - mean(x) ** 2
```

```python
def std(x):
    return var(x) ** 0.5
```

```python
mean(x), mean(y)
```

```python
var(x), var(y)
```

```python
std(x), std(y)
```

```python
def cov(x, y):
    # return ((x - mean(x)) * (y - mean(y))).sum() / (len(x) - 1)
    return mean(x * y) - mean(x) * mean(y)
```

```python
cov(x, y)
```

```python
def corr(x, y):
    return cov(x, y) / (std(x) * std(y))
```

```python
corr(x, y)
```

```python
import numpy as np
from scipy import stats

# 1. Define the sample data
sample_weights = [49.2, 50.1, 50.3, 49.8, 50.5]
hypothesized_mean = 50
alpha = 0.05 # Significance level

# 2. Define the Hypotheses
# Null Hypothesis (H0): The true average weight is 50 grams.
# Alternative Hypothesis (H1): The true average weight is not 50 grams.

# 3. Perform the one-sample t-test
t_stat, p_value = stats.ttest_1samp(sample_weights, hypothesized_mean)

# 4. Interpret the results
print(f"Sample data: {sample_weights}")
print(f"Hypothesized mean: {hypothesized_mean} grams")
print(f"T-statistic: {round(t_stat, 4)}")
print(f"P-value: {round(p_value, 4)}")

if p_value <= alpha:
    print("\nConclusion: The null hypothesis is rejected.")
    print("The average weight is significantly different from 50 grams.")
else:
    print("\nConclusion: The null hypothesis is not rejected.")
    print("There is no significant difference from 50 grams.")
```
